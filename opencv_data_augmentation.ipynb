```python
# edge detection & image gradients
img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_01.jpeg',0)

height,width=img.shape

# extract sobel edges
sobel_x=cv2.Sobel(img,cv2.CV_64F,0,1, ksize=5)
sobel_y=cv2.Sobel(img,cv2.CV_64F,1,0, ksize=5)

cv2.imshow('Original', img)
cv2.waitKey(0)
cv2.imshow('Sobel x', sobel_x)
cv2.waitKey(0)
cv2.imshow('Sobel y', sobel_y)
cv2.waitKey(0)

sobel_OR=cv2.bitwise_or(sobel_x,sobel_y)
cv2.imshow('Sobel OR', sobel_OR)
cv2.waitKey(0)

laplacian=cv2.Laplacian(img,cv2.CV_64F)
cv2.imshow('Laplacian', laplacian)
cv2.waitKey(0)

canny=cv2.Canny(img,50,120)
cv2.imshow('Canny', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# transelations of image 

img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_01.jpeg',0)

# store height and width of the image 
height,width=img.shape[:2]

quarter_height, quarter_width = height/4, width/4

# T is our transelation matrix
T=np.float32([[1,0,quarter_width],[0,1,quarter_height]])

# we use warpaffine to transform the image using the matrix T

img_transelation=cv2.warpAffine(img,T,(width,height))
cv2.imshow('Transelationy', img_transelation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# rotations
img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_01.jpeg')

# store height and width of the image 
height,width=img.shape[:2]

# devided by 2 to rotate the image around its centre
rotation_matrix=cv2.getRotationMatrix2D((width/2,height/2),30,.75)

rotated_image=cv2.warpAffine(img,rotation_matrix,(width,height))
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Horizontal flip

flipped=cv2.flip(img,1)
cv2.imshow('Horizontal Flip', flipped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# arithmetic operations in Brightness

img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_02.jpeg')

# creating a matrix of ones then mmultiply it by a scaler of 100 
# this gives a matrix with same dimensions of our image with all values being 100
B=np.ones(img.shape, dtype = "uint8")*75
cv2.imshow("Original", img)

# use this to add this matrix B to our img and increasing in brightness
added=cv2.add(img,B)
cv2.imshow("Added", added)

# likewise we can subtract, decrease in brightness
subract=cv2.subtract(img,B)
cv2.imshow("Subract", subract)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# sharpening we can implement by altering kernal

img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_02.jpeg')
cv2.imshow("Original", img)

kernal_sharpening=np.array([[-1,-1,-1],
                           [-1,11,-1],
                           [-1,-1,-1]])
# applying diff kernal to the input image
sharped=cv2.filter2D(img,-1,kernal_sharpening)
cv2.imshow("Img Sharped", sharped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Blurring

img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_02.jpeg')
cv2.imshow("Original", img)
cv2.waitKey(0)

# creating our 3*3 kernal
kernal_3x3=np.ones((3,3),np.float32)/9

# we use the cv2.filter2D to convolve the kernal with an image
blurred=cv2.filter2D(img,-1,kernal_3x3)
cv2.imshow("3x3 kernal Blurring", blurred)
cv2.waitKey(0)

# creating our 7*7 kernal
kernal_7x7=np.ones((7,7),np.float32)/49

blurred=cv2.filter2D(img,-1,kernal_7x7)
cv2.imshow("7x7 kernal Blurring", blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('C:/Users/rishi.yadav/Web Scraping/Flipcart_Mens_trouser_jeans_dataset/single_prediction/Mens_jeans_or_trouser_02.jpeg')
cv2.imshow("Original", img)
cv2.waitKey(0)

# averaging done by convolving the image  with a normalized box filter
blur=cv2.blur(img,(3,3))
cv2.imshow("Averaging", blur)
cv2.waitKey(0)

# instead of box filter gaussain kernal
gaussain=cv2.GaussianBlur(img,(7,7),0)
cv2.imshow("Gaussain", gaussain)
cv2.waitKey(0)

# instead of box filter median kernal
median=cv2.medianBlur(img,5)
cv2.imshow("Median", median)
cv2.waitKey(0)

# Bilateral is very effective in noice removal while keeping edges sharp
bilateral=cv2.bilateralFilter(img,9,75,75)
cv2.imshow("Bilateral Blur", bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
