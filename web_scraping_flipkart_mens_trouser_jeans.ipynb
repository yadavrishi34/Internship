```python
# step 1 import the libraies and get url of the web page
from selenium import webdriver
import os
```

```python
driver=webdriver.Chrome(executable_path='C:\driver\chromedriver')
```

```python
# print page 1 of flip cart mens trouser
my_page=driver.get("https://www.flipkart.com/search?q=Men%27s+Trousers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY")
```

```python
# step 2 making the directory to store images
def make_directory(dirname):
    current_path=os.getcwd()
    path=os.path.join(current_path,dirname)
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
make_directory("flipcart_trousers")
```

```python
# step 3 scraping image urls
images=driver.find_elements_by_xpath("//img[@class='_3togXc']")
print(images[0])
```

```python
urls=[]
for image in images:
    source=image.get_attribute('src')
    urls.append(source)
    print(source)
```

```python
len(urls)
```

```python
# step 4 using the urls to download images
import shutil
import requests
```

```python
for index, link in enumerate(urls):
    print("Downloading {0} of {1} images".format(index+1,len(urls)))
    response=requests.get(link)
    with open('flipcart_trousers/img_{0}.jpeg'.format(index),"wb") as file:
        file.write(response.content)
```
